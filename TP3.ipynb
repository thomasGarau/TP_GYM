```python
import gym
import pandas as pd
import numpy as np
import random 
from collections import defaultdict
from collections import defaultdict
env = gym.make('Blackjack-v1')
env.reset()
```

```python
env.action_space
```

```python
def policy(state):
    score, dealer_score, as_enable = state
    return 0 if score > 19 else 1

def generate_episode():
    episode = []
    state, info = env.reset()
    while True:
        action = policy(state)
        next_state, reward, done, info, com = env.step(action)
        episode.append((state, action, reward))
        if done:
            return episode
        state = next_state


print(generate_episode())
    
    
```

```python
nombre_iteration = 50000
def value_function():
    Total_return = defaultdict(float)
    N = defaultdict(float)
    V = defaultdict(float)
    for i in range(nombre_iteration):
        episode = generate_episode()
        for j, (state,_,_) in enumerate(episode):
            Total_return[state] += sum(reward[2] for reward in episode[j:])
            N[state] += 1

    for state in Total_return:
        V[state] = Total_return[state] / N[state]

    #sorted retourne la liste des clées dans l'ordre croissant
    sorted_keys = sorted(V.keys())
    #On crée à nouveau un dico en itérant sur la liste de clés triée
    return {key: V[key] for key in sorted_keys}
value_function()
        
    
```

```python
def monte_carlo(nb_iter):
    Total_return = defaultdict(lambda: defaultdict(float))
    N = defaultdict(lambda: defaultdict(float))
    V = defaultdict(float)

    for i in range(nb_iter):
        episode = generate_episode()
        for j, (state, action, _) in enumerate(episode):
            # s'il y a deja eu cette action dans cet état, on ne la compte pas car on ne veut que la première visite
            if (action not in N[state]):
                Total_return[state][action] += sum(reward[2] for reward in episode[j:])
                N[state][action] += 1

    sorted_keys = sorted(Total_return.keys())
    return {key: Total_return[key] for key in sorted_keys}


# Exemple d'appel de fonction
monte_carlo(50000)
```

```python
#On crée à nouveau ces fonctions car on a besoin d'une policy qui permettra d'obtenir toutes les q values pour toutes les actions possibles 
#Or avec la policy précedente pour un même state on utilise toujours le même état le tableau de q_value est donc à moitié vide 
#On code à nouveau monte_carlo en ne se limitant pas à la première visite pour être plus précis

def random_policy():
    return np.random.randint(0,2)

def generate_random_episode():
    episode = []
    state, info = env.reset()
    while True:
        action = random_policy()
        next_state, reward, done, info, com = env.step(action)
        episode.append((state, action, reward))
        if done:
            return episode
        state = next_state

def monte_carlo(nb_iter):
    Total_return = defaultdict(lambda: defaultdict(float))
    N = defaultdict(lambda: defaultdict(float))
    V = defaultdict(lambda: defaultdict(float))

    for i in range(nb_iter):
        episode = generate_random_episode()
        for j, (state, action, _) in enumerate(episode):
            #S'il y a deja eu cette action dans cet état, on ne la compte pas car on ne veut que la première visite
            #if (action not in N[state]):
                Total_return[state][action] += sum(reward[2] for reward in episode[j:])
                N[state][action] += 1

    for state, actions in Total_return.items():
        for action in actions:
            V[state][action] = Total_return[state][action] / N[state][action]

    # Tri et retour du dictionnaire
    sorted_keys = sorted(V.keys())
    return {key: V[key] for key in sorted_keys}


# Exemple d'appel de fonction
monte_carlo(50000)
```

```python
def epsilon_greedy_policy(q_value, state, epsilon):
    # Vérifie si l'état est dans q_value et si les actions pour cet état existent
    if state not in q_value or 0 not in q_value[state] or 1 not in q_value[state] or np.random.random() < epsilon:
        return np.random.randint(0, 2)  # Choix aléatoire
    else:
        return np.argmax([q_value[state][0], q_value[state][1]])  # Choix de l'action avec le Q le plus élevé

q_value = monte_carlo(50000)
epsilon_greedy_policy(q_value, (4, 5, False), 0.1)
```

```python
def generate_policy_episode(q_value,epsilon):
    episode = []
    state, info = env.reset()
    while True:
        action = epsilon_greedy_policy(q_value, state, epsilon)
        next_state, reward, done, info, com = env.step(action)
        episode.append((state, action, reward))
        if done:
            return episode
        state = next_state
```

```python
def update_q_value(q_values, episode, alpha):

    for i, (state, action, reward) in enumerate(episode):
        # Calcul du retour (somme des récompenses futures) à partir de cette étape
        G = sum(x[2] for x in episode[i:])
        # Mise à jour de la valeur Q pour l'état et l'action donné
        # Q(s,a) = Q(s,a) + alpha * (G - Q(s,a))
        if state in q_values and action in q_values[state]:
            q_values[state][action] += alpha * (G - q_values[state][action])
        else:
            # Initialisation si la paire état-action n'est pas encore dans q_values
            q_values[state][action] = G  
    return q_values

#update la q value pour obtenir une politique plus optimale 
q_value = monte_carlo(50000)
for i in range(500000):
    q_value = update_q_value(q_value, generate_policy_episode(q_value,0.1), 0.1)
q_value
   	
```

```python
import copy

def has_converged(old_q_values, new_q_values, threshold):
    total_diff = 0
    count = 0
    for state in new_q_values:
        for action in new_q_values[state]:
            # si le state n'existait pas avant cette série d'intération on ne le compte pas
            old_value = old_q_values[state][action] if state in old_q_values and action in old_q_values[state] else 0
            total_diff += abs(new_q_values[state][action] - old_value)
            count += 1
    average_diff = total_diff / count if count > 0 else 0
    return average_diff < threshold

def find_optimal_policy():
    # Seuil de convergence
    threshold = 0.001
    #Taux d'apprentissage (impact qu'aura le nouvel épisode)
    alpha = 0.1
    #Nombre d'itération maximun pour essayer d'atteindre la convergence * 1000
    nb_iter = 1000
    # Initialisation des valeurs Q
    old_q_values = monte_carlo(50000)
    #On utilise un deepcopy pour ne pas modifier old_q_values car copy simple ne fait que copier la référence (adresse mémoire)
    new_q_values = copy.deepcopy(old_q_values)
    for _ in range(nb_iter):
        #Toutes les milles épisodes générés on test si la policy a changé de manière significative ou non
        for _ in range(1000):
            episode = generate_policy_episode(new_q_values, 0.1)
            new_q_values = update_q_value(new_q_values, episode, alpha)
        if has_converged(old_q_values, new_q_values, threshold):
            return new_q_values
        old_q_values = copy.deepcopy(new_q_values)
    #Si la convergence n'est pas atteinte après un million d'itération on arrête et on retourne la policy actuelle
    return new_q_values
```

On constate qu'il est impossible de trouver le point de convergence ou la policy est optimal car elle se met à jour constament. Etant donné que les valeurs des rewards sont extremes -1 +1 par rapport à celle des q value qui sont toujours comprises dans cette intervalle, à chaque ajout d'épisode la table de q value est donc toujours modifiée un peu. On constate également que le taux de variation de chaque 1000 update de la q table est très corréler de celui du taux d'apprentissage alpha.
